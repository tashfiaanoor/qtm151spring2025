```python
import matplotlib.pyplot as plt

# Data for the requests
categories = ['Education', 'Climate', 'Livestock', 'Health', 'Finance', 'Agriculture', 'Training', 'Govt Operations', 'Social Benefits/Allowances']
requests = [25, 15, 10, 20, 5, 12, 8, 3, 2]  # Example data: Total requests for each category

# Define the colors for each category
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c4e17f','#8fc0ff','#bfa3b8']

# Plot the pie chart
plt.figure(figsize=(8,8))
plt.pie(requests, labels=categories, colors=colors, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Proportion of Requests by Category')

# Add a legend with the colors and their corresponding categories
plt.legend(categories, title="Request Categories", loc="upper left", bbox_to_anchor=(1, 1))

# Display the pie chart
plt.show()
```
